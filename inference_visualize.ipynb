```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
import numpy as np

# csv 파일 읽기
df = pd.read_csv('/opt/ml/DACON-objectdetection/results/sy_default_epoch15.csv')
data_root = '/opt/ml/DACON-objectdetection/data/test/'
```

```python
# unique file_names
unique_files = df.file_name.unique()

target_files = np.random.choice(unique_files, 1)

# loop over each unique file
for file in target_files:
    # filter the data for the current image
    current_data = df[df.file_name == file]
    file = data_root+file

    # 이미지 불러오기
    img = cv2.imread(file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # create a plot
    fig, ax = plt.subplots(1)
    ax.imshow(img)

    # loop over the bounding boxes and draw them
    for idx, row in current_data.iterrows():
        # bounding box coordinates
        box = [(row.point1_x, row.point1_y), 
               (row.point2_x, row.point2_y),
               (row.point3_x, row.point3_y), 
               (row.point4_x, row.point4_y)]

        # Create a Rectangle patch
        rect = patches.Polygon(box, linewidth=1, edgecolor='r', facecolor='none')

        # Add the patch to the Axes
        ax.add_patch(rect)
        # plt.text(row.point1_x, row.point1_y, f"{row.class_id}:{row.confidence}", color='r')
        plt.text(row.point1_x, row.point1_y, '%d : %.3f'%(row.class_id, row.confidence), color='r')
        print('%d : %.3f'%(row.class_id, row.confidence))

    

    plt.show()
```
